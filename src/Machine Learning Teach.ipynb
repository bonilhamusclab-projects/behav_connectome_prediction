## How Cross Validation Works

```julia
sample_ixs = 1:5
num_samples = length(sample_ixs)
train_size=4
num_iterations = 6

for ixs in RandomSub(num_samples, train_size, num_iterations)
    println("***")
    
    println("train indexes: $ixs")
    
    test_indexes = setdiff(sample_ixs, ixs)
    println("test indexes: $test_indexes")
    
    println("###")
end
```

### Cross Validate Helper Function

```julia
@doc(cross_validate)
```

```julia
call_state = Dict(:ix => 0)

function fit(inds::Vector{Int64})
    call_state[:ix] += 1
    call_ix = call_state[:ix]
    println("****")
    println("fit: call $call_ix with $inds")
end

function test(_, inds::Vector{Int64})
    call_ix = call_state[:ix]
    println("test: call $call_ix with $inds")
    println("###")
    score = .9
    score
end
```

```julia
cvg = RandomSub(num_samples, train_size, num_iterations)
cross_validate(fit, test, num_samples, cvg)
```

## Can We Do Better?

### Let's look at the regularization param

__Evaluate Train Scores With Test Scores__

```julia
function eval_model(C::Float64)
    svr = LinearSVR(C=C)
    srand(1234)
    
    num_iterations = 100
    num_samples = length(diff_wpm_scores)
    
    cvg = RandomSub(num_samples, round(Int64, .8 * num_samples), num_iterations)

    train_scores = zeros(Float64, num_iterations)
    state = Dict(:fit_call => 0)

    function predict_score(inds::Vector{Int64})
        predictions = svr[:predict](X_lesion[inds, :])
        truths = diff_wpm_scores[inds]
        r2_score(truths, predictions)
    end

    function fit(inds::Vector{Int64})
        svr[:fit](X_lesion[inds, :], diff_wpm_scores[inds])

        state[:fit_call] += 1
        train_scores[state[:fit_call]] = predict_score(inds)
    end

    test(_, inds::Vector{Int64}) = predict_score(inds)

    test_scores = cross_validate(fit, test, num_samples, cvg)
    train_scores, test_scores
end
```

```julia
train_scores, test_scores = eval_model(1.);
```

```julia
println("train $(mean(train_scores)) vs test $(mean(test_scores))")
```

__Clear Underfitting, Try new C Values__

```julia
possible_Cs = 1.:5:41.

train_score_mns = zeros(Float64, length(possible_Cs))
test_score_mns = zeros(Float64, length(possible_Cs))

train_score_stds = zeros(Float64, length(possible_Cs))
test_score_stds = zeros(Float64, length(possible_Cs))

for (ix, c) in enumerate(possible_Cs)
    svr[:C] = c
    train_scores, test_scores = eval_model(c)
    train_score_mns[ix] = mean(train_scores)
    test_score_mns[ix] = mean(test_scores)
    
    train_score_stds[ix] = std(train_scores)
    test_score_stds[ix] = std(test_scores)
end
```

```julia
using Colors
using DataFrames
using Gadfly

plot_info = DataFrame(train_mn=train_score_mns, test_mn=test_score_mns, C=possible_Cs) 
plot_info[:legend] = "train"
plot_info[:test_color] = "test"

mk_layer(y, color) = layer(x="C", y=y, Geom.point, Geom.line, color=color)

plot(plot_info, mk_layer("train_mn", "legend"), mk_layer("test_mn", "test_color"),
Guide.xlabel("C"), Guide.ylabel("mean R2"))
```
